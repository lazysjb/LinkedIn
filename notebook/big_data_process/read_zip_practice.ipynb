```python
def _is_dir(filename):
    """Heuristic"""
    return filename.endswith('/')

def _get_dir_from_filename(filename):
    return filename.split('/')[0]
```

```python
def get_directories(zf):
    directories = []
    
    for filename in zf.namelist():
        if _is_dir(filename):
            directory = _get_dir_from_filename(filename)
            directories.append(directory)
    
    return directories

def get_filenames_for_directory(zf, usr_directory):
    filenames = []
    
    for filename in zf.namelist():
        if not _is_dir(filename):
            directory = _get_dir_from_filename(filename)
            if directory == usr_directory:
                filenames.append(filename)
    
    return filenames

def get_file_content_generator_from_directory(zf, usr_directory):
    for filename in zf.namelist():
        if not _is_dir(filename):
            directory = _get_dir_from_filename(filename)
            if directory == usr_directory:
                with zf.open(filename) as f:
                    content = f.read()
                    yield content
```
